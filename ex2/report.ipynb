### **Intrepretao do enunciado:**
 desejado que se implemente uma classe em python que represente um EcDSA.

Um EcDSA  "Elliptic Curve Digital Signature Algorithm" ou seja  desejado que seja implementado um Algoritmo de Assinatura Digital recorrendo a Curvas Elpticas.

 desejado que este algoritmo seja implementado a partir do standard *FIPS186-5*, que especifica o Digital Signature Standard (DSS), ou seja, contm a especificao dos algoritmos que devem ser utilizados para implementar assinaturas digitais.  quais as curvas elipticas que devem ser utilizadas na implementao de um EcDSA.

#### **Analise do FIPS186-5**
Aps uma leitura do FIPS186-5  possivel destacar que os algoritmos de hash utilizados na implementao dos algoritmos de assinatura digital devem utilizar funes de hash aprovadas, algumas delas so o FIPS180 e o FIPS202.




Os parametros utilizados para gerar a assinatura devem seguir o SP 800-89 e/ou SP 800-186.



### **Escolha da curva:**
Escolhi implementar
A implementao escolhida ir utilizar Ed448 pois oferece aproximadamente uma segurana de 224-bits ao invs de 128-bits de segurana do Ed25519.
